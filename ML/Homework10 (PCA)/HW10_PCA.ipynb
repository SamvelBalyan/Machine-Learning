```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets

plt.style.use('seaborn-whitegrid') # Plot style

%load_ext autoreload
%autoreload 2
```

We will use the Iris dataset to experiment with PCA.

```python
iris = sklearn.datasets.load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris.head()
```

```python
from pca import PCA
model = PCA(nr_components=2)
X_reduced = model.fit_transform(df_iris)
```

You should get about $97.77\%$ explained variance with $2$ principal components and a graph like the below one if you execute the code below.

```python
plt.plot(X_reduced)
```

```python
print('Explained variance:', model.explained_variance_ratio)

colors = np.array(['red', 'blue', 'green'])

plt.scatter(X_reduced.iloc[:,0], X_reduced.iloc[:,1],
            color=colors[iris.target]);
```

```python
from sklearn.decomposition import PCA
```

```python
model2 = PCA(n_components=2)
X_reduced2 = model2.fit_transform(df_iris)
```

```python
print('Explained variance:', model2.explained_variance_ratio_)

colors = np.array(['red', 'blue', 'green'])

plt.scatter(X_reduced2[:,0], X_reduced2[:,1],
            color=colors[iris.target]);
```
