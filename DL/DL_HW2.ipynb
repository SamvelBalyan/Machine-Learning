```python
#stochastic GD

import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize = (10,10))

X = np.random.rand(1000)
y = (1+X)**0.5
w = 1
b = 1
def dL2_dw(w,b,i):
  return (w*X[i]+b-y[i])*X[i]/500
def dL2_db(w,b,i):
  return (w*X[i]+b-y[i])/500
a = 0.1
for epoch in range(1000):
  for i in range(len(X)):
    w = w - a*dL2_dw(w,b,i)
    b = b - a*dL2_db(w,b,i)


print(w,b)
plt.scatter(X,y)
plt.plot(X,w*X+b,'r')
plt.show()
```

```python
# mini batch GD

plt.figure(figsize = (10,10))

x = np.random.rand(1000,1)
y = (1+x)**0.5
b = np.ones((1000,1))

X = np.concatenate([x,b],axis=1)
w = np.array([3.,3.])
batchsize = 10
def dl_dw(w,i):
  func = 0
  for j in range(i,len(X),int(len(X)/batchsize)):
    func += (X[j]@w-y[j])*X[j]
  func /= 500
  return func
a = 0.5
for epoch in range(150):
  for i in range(int(len(X)/batchsize)):
    w -= a*dl_dw(w,i)
    
plt.scatter(x,y)
plt.plot(x,X@w,'r')
plt.show()

print(w)

```
